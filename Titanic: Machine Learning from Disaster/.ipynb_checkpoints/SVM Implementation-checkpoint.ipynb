```python
''' Import libaries '''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
''' Display the data '''

# Open train.csv and display the first 10 rows. We are going to first drop all passengers with NaN data points
# and plot the resulting features to get an idea of how the data look
train = pd.read_csv('train.csv')
train.head()
```

```python
''' Visualization of features and distribution of survival rate for each feature.'''

def plot_data(train_clean):
    grid = GridSpec(2, 4)
    plt.subplots_adjust(left=0.1, bottom=0.1, right=2, top=2, wspace=0.5, hspace=None)

    # Display survival rate for each sex
    N = 2
    sex = train_clean['Sex'].copy()
    male_idx = sex[train_clean['Sex'] == 'male'].index
    female_idx = sex[train_clean['Sex'] == 'female'].index
    survival = (sum(train_clean['Survived'][male_idx] == 1),sum(train_clean['Survived'][female_idx] == 1))
    dead = (sum(train_clean['Survived'][male_idx] == 0),sum(train_clean['Survived'][female_idx] == 0))
    ind = np.arange(N)
    width = 0.6

    plt.subplot(grid[0,0])
    p1 = plt.bar(ind,survival,width,alpha=0.7,color='r')
    p2 = plt.bar(ind,dead,width,alpha=0.7,color='c',bottom=survival)

    plt.ylabel('Passengers')
    plt.title('Survival by sex')
    plt.xticks(ind,('Male','Female'))
    plt.legend((p1[0],p2[0]),('Survival','Dead'))


    # Display survival rate for each class
    N = 3 
    pclass = train['Pclass'].copy()
    first = pclass[train['Pclass'] == 1].index
    second = pclass[train['Pclass'] == 2].index
    third = pclass[train['Pclass'] == 3].index
    survival = (sum(train['Survived'][first] == 1),sum(train['Survived'][second] == 1),
                sum(train['Survived'][third] == 1))
    dead = (sum(train['Survived'][first] == 0),sum(train['Survived'][second] == 0),
                sum(train['Survived'][third] == 0))
    ind = np.arange(N)
    width = 0.8

    plt.subplot(grid[0,1])
    p1 = plt.bar(ind,survival,width,alpha=0.7,color='r')
    p2 = plt.bar(ind,dead,width,alpha=0.7,color='c',bottom=survival)

    plt.ylabel('Passengers')
    plt.title('Survival by ticket class')
    plt.xticks(ind,('1st','2nd','3rd'))
    plt.legend((p1[0],p2[0]),('Survival','Dead'))


    # Display survival rate for each embarked port
    N = 3
    depart = train['Embarked'].copy()
    cherbourg = depart[train['Embarked'] == 'C'].index
    queenstown = depart[train['Embarked'] == 'Q'].index
    southampton = depart[train['Embarked'] == 'S'].index
    survival = (sum(train['Survived'][cherbourg] == 1),sum(train['Survived'][queenstown] == 1),
                sum(train['Survived'][southampton] == 1))
    dead = (sum(train['Survived'][cherbourg] == 0),sum(train['Survived'][queenstown] == 0),
                sum(train['Survived'][southampton] == 0))
    ind = np.arange(N)
    width = 0.8

    plt.subplot(grid[0,2])
    p1 = plt.bar(ind,survival,width,alpha=0.7,color='r')
    p2 = plt.bar(ind,dead,width,alpha=0.7,color='c',bottom=survival)

    plt.ylabel('Passengers')
    plt.title('Survival by departure port')
    plt.xticks(ind,('C','Q','S'))
    plt.legend((p1[0],p2[0]),('Survival','Dead'))


    # Plot distribution of ages for survival/non-survival. Eliminate NaNs first.
    age_survive = train_clean['Age'][train_clean.index[train_clean['Survived'] == 1]]
    age_dead = train_clean['Age'][train_clean.index[train_clean['Survived'] == 0]]
    plt.subplot(grid[1,0:2])
    plt.hist([age_survive,age_dead],stacked = True,color=['r','c'],edgecolor ='black',
             alpha =0.75, bins = 40,label=['Survived','Dead'])
    plt.title('Age distributions for survival/dead')
    plt.xlabel('Age (years)')
    plt.ylabel('N')
    plt.legend()


    # Plot distribution of fares for survival/non-survival. Eliminate NaNs firstt
    fare_survive = train_clean['Fare'][train_clean.index[train_clean['Survived'] == 1]]
    fare_dead = train_clean['Fare'][train_clean.index[train_clean['Survived'] == 0]]
    plt.subplot(grid[1,3:3])
    plt.hist([fare_survive,fare_dead],stacked = True, color=['r','c'],edgecolor = 'black',
             alpha = 0.75,bins = 40,label=['Survived','Dead'])
    plt.title('Fare distributions for survival/dead ')
    plt.xlabel('Fare ($)')
    plt.ylabel('N')
    plt.legend()




    # Plot distribution of siblings/spouses
    sibsp = train_clean['SibSp']

    # Plot distribution of parents/children

    
    
    plt.show()
    
    return None 

plot_data(train_clean)
```

```python
'''Feature extraction and scaling'''

def extract_scale(train):
    # Feature extraction and scaling
    # Note: For now we will ignore the ticket number feature and the name feature.
    # Normalize ages
    age = train['Age']
    age_norm = (age - np.mean(age))/np.std(age)

    # Normalize fares
    fare = train['Fare']
    fare_norm = (fare - np.mean(fare))/np.std(fare)

    # Normalize sibling-spouses
    sibsp = train['SibSp']
    sibsp_norm = (sibsp - np.mean(sibsp))/np.std(sibsp)

    # Normalize parents-childrens
    parch = train['Parch']
    parch_norm = (parch - np.mean(parch))/np.std(parch)

    # Binarize sex
    sex = train_clean['Sex']
    sex_bin = [1 if i == 'female' else 0 for i in sex]

    # Binarize ticket class
    pclass = train['Pclass']
    first_class = [1 if i == 1 else 0 for i in pclass]
    second_class = [1 if i == 2 else 0 for i in pclass]
    third_class = [1 if i == 3 else 0 for i in pclass]

    # Binarize departure port
    depart = train['Embarked']
    cherbourg = [1 if i == 'C' else 0 for i in depart]
    queenstown = [1 if i == 'Q' else 0 for i in depart]
    southampton = [1 if i == 'S' else 0 for i in depart]

    # Binarize whether or not cabin number was reported (reported - 1; not reported - 0)
    cabin = [0 if pd.isnull(i) is True else 1 for i in train['Cabin']]
    
    # Process and binarize titles
    titles = {'Mrs.':     'Missus',
              'Mr.':      'Mister',
              'Miss.':    'Miss',
              'Ms.':      'Miss',
              'Master.':  'Master',
              'Don.':     'Mister',
              'Rev.':     'Reverend',
              'Dr.':      'Doctor',
              'Mme.':     'Madame',
              'Major.':   'Officer'
              'Lady.':    'Lady',
              'Sir.':     'Sir',
              'Mlle.':    'Madamoiselle',
              'Col.':     'Officer',
              'Capt.':    'Captain',
              'Countess': 'Royalty'}
    
    
    data = dict()    
    # Construct training matrix as X
    data['X'] = np.transpose(np.array([first_class, 
                                       second_class, 
                                       third_class, 
                                       sex_bin, 
                                       age_norm, 
                                       sibsp_norm,
                                       parch_norm, 
                                       fare_norm, 
                                       cabin, 
                                       cherbourg, 
                                       queenstown, 
                                       southampton]))
    
    # Construct response vector as Y
    try:
        print('---Extracting response variables----')
        data['Y'] = np.transpose(np.array(train['Survived']))
        
    except KeyError:
        print('----Test set has no response variables----')
        data['Y'] = ['unused']
        
    return scaled_data

scaled_data = extract_scale(processed_data)
```
